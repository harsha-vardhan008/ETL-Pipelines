```python
def update_zip_code(df, customer_id, new_zip):
    idx = df[df['customer_id'] == customer_id].index[0]
    current_new = df.at[idx, 'new_zip_code']
   
    if not pd.isna(current_new):
        df.at[idx, 'previous_zip_code'] = current_new
    else:
        df.at[idx, 'previous_zip_code'] = df.at[idx, 'zip_code']
    df.at[idx, 'new_zip_code'] = new_zip
 
    return df
 
```

```python
import pandas as pd
from datetime import datetime
```

```python
# Creating new table
customer_history = pd.DataFrame(columns=[
    'history_id', 'customer_id', 'first_name', 'last_name', 'email',
    'city', 'state', 'zip_code', 'active_or_inactive', 'change_date'
])
```

```python
import pandas as pd
from datetime import datetime

def apply_scd_type4(df, customer_history, incoming):
    
    #  Check if customer exists
    if df[df['customer_id'] == incoming['customer_id']].empty:
        print("Customer not found ")
        return df, customer_history

    # Get current record
    current = df[df['customer_id'] == incoming['customer_id']].iloc[0]

    #  Fields to compare 
    fields_to_compare = ['first_name', 'last_name', 'email', 'city', 'state', 'zip_code', 'active_or_inactive']

    #  Check for change
    has_change = any(current[field] != incoming[field] for field in fields_to_compare)

    if has_change:
        # Move old record to history table
        history_entry = current.to_dict()
        history_entry['history_id'] = 1 if customer_history.empty else customer_history['history_id'].max() + 1
        history_entry['change_date'] = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

        customer_history = pd.concat([customer_history, pd.DataFrame([history_entry])], ignore_index=True)

        # Update current dimension table
        df.loc[df['customer_id'] == incoming['customer_id'], fields_to_compare] = incoming[fields_to_compare].values

    return df, customer_history
```

```python
incoming = pd.Series({
    'customer_id': 4,
    'first_name': 'Jamie',
    'last_name': 'Smith',
    'email': 'john.new@xyz.com',
    'city': 'Boston',
    'state': 'LA',
    'zip_code': '02118',
    'active_or_inactive': 'active'
})
#calling the function
df, customer_history = apply_scd_type4(df, customer_history, incoming)
```

# SCD TYPE 5

```python
import pandas as pd
from datetime import datetime

def apply_scd_type5(df, customer_historys, incoming):
    
    #  Check if customer exists
    if df[df['customer_id'] == incoming['customer_id']].empty:
        print("Customer not found")
        return df, customer_historys

    # Get current record
    current = df[df['customer_id'] == incoming['customer_id']].iloc[0]

    # Define fields
    overwrite_fields = ['first_name', 'last_name', 'email', 'city', 'state']
    history_fields = ['zip_code', 'active_or_inactive']

    # Check if any history field changed
    has_history_change = any(current[field] != incoming[field] for field in history_fields)

    if has_history_change:
        # Move current record to history
        history_entry = current.to_dict()
        history_entry['history_id'] = 1 if customer_historys.empty else customer_historys['history_id'].max() + 1
        history_entry['change_date'] = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
        customer_historys = pd.concat([customer_historys, pd.DataFrame([history_entry])], ignore_index=True)

    # Overwrite all changed fields (both overwrite_fields and history_fields)
    for field in overwrite_fields + history_fields:
        if current[field] != incoming[field]:
            df.loc[df['customer_id'] == incoming['customer_id'], field] = incoming[field]

    return df, customer_historys
```

```python
df = pd.read_csv("customer.csv")
```

```python

# Initially empty history table
customer_historys = pd.DataFrame(columns=[
    'history_id', 'customer_id', 'first_name', 'last_name', 'email',
    'city', 'state', 'zip_code', 'active_or_inactive', 'change_date'
])

# Incoming update
incoming = pd.Series({
    'customer_id': 10,
    'first_name': 'Pamelia',
    'last_name': 'Harsha',
    'email': 'pamelia.newman@gmail.com',
    'city': 'Monroe',
    'state': 'NY',
    'zip_code': '10950',  # zip_code changed (should trigger history)
    'active_or_inactive': 'active'  # also changed
})


df, customer_historys = apply_scd_type5(df, customer_historys, incoming)

```
